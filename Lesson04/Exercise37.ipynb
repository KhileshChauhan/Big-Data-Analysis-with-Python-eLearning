```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("Python Spark Session").getOrCreate()
df = spark.read.format('com.databricks.spark.csv').options(header='true', inferschema='true').load('iris.csv')
df.show(4)
```

```python
import seaborn as sns
sns.set()
data = df.toPandas()
sns.kdeplot(data.sepal_width, shade = True)
plt.show()
```

```python
sns.boxplot(x = "sepal_length", y = "sepal_width", data = data)
plt.show()
```
